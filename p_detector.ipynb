```python
import numpy as np  
import pandas as pd
df = pd.read_csv("Phishing_Dataset.csv") #import dataset
df.describe()
```

```python
y = df["phishing"] # stores phishing variable into df called y
df.drop(["phishing"], axis = 1, inplace=True) # drops the only y variable from the dataset
df.describe()
```

```python
# removing values in df that are negative or missing)
df[df < 0] = 0 # Replace negative values in df with 0 (in-place
df.describe()
```

```python
# replace missing values in df with 0
df.fillna(0, inplace=True)
df.describe()
```

```python
# Identify and drop columns in df with variance less than or equal to 0.5
low_variance = []  # List to store columns with low variance
counter = 0  # Counter for low variance columns
for col in df.columns:
    # Check if the variance of the column is less than or equal to 0.5
    if df[col].var() <= 0.5:
        low_variance.append(col)  # Add column to the list
        counter += 1  # Increment counter
print(f"Number of low variance columns: {counter}")  # Print the number of low variance columns

# Drop the low variance columns from df
df.drop(columns=low_variance, inplace=True)
df.describe()  # Show summary statistics of the updated DataFrame
```

```python
# split into training and testing sets

# compare decision, neural network, and logistic regression models
# store in pkl files for streamlit app

# Evaluate the model
#print("Accuracy:", accuracy_score(y_test, y_pred))
#print("Classification Report:\n", classification_report(y_test, y_pred))
# from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
```

```python
# Import classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initialize models
dt_model = DecisionTreeClassifier(random_state=42)
nn_model = MLPClassifier(random_state=42, max_iter=1000)
lr_model = LogisticRegression(max_iter=1000, random_state=42)
```

```python
# Train models
dt_model.fit(X_train, y_train)
nn_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred_dt = dt_model.predict(X_test)
y_pred_nn = nn_model.predict(X_test)
y_pred_lr = lr_model.predict(X_test)
```

```python
# Compare accuracy
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Neural Network Accuracy:", accuracy_score(y_test, y_pred_nn))
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
# Save the trained Decision Tree model to a pickle file
import pickle
with open('decision_tree_model.pkl', 'wb') as f:
    pickle.dump(dt_model, f)
print("Decision Tree model saved as decision_tree_model.pkl")
```

```python
import pickle
with open('neural_network_model.pkl', 'wb') as f:
    pickle.dump(nn_model, f)
print("Neural Network model saved as neural_network_model.pkl")
```

```python
with open('log_reg_model.pkl', 'wb') as f:
    pickle.dump(lr_model, f)
print("Logistic Regression model saved as log_reg_model.pkl")
```
